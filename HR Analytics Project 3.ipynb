# HR Employee Attrition Visualization

```python
hr_employee_data["Attrition"].value_counts().plot(kind = 'density', color = 'g')
print(hr_employee_data.Attrition.value_counts())
plt.xlabel('Attrition', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
```

```python
# crosschecking for the values contained in each column

for col in hr_employee_data.describe(include = 'object').columns:
    print(col)
    print(hr_employee_data[col].unique())
    print(','*40)
```

```python
hr_cols = hr_employee_data.select_dtypes(['object']).columns

hr_cols
```

```python
# Plotting the pie chart for column distributions

plotnumber = 1
plt.figure(figsize =(30,24), facecolor = 'white')
for col in hr_cols:
    if(plotnumber <=9):
        plt.subplot(4,4,plotnumber)
        plt.pie(hr_employee_data[col].value_counts(), labels = hr_employee_data[col].value_counts().values)
        plt.title(col+"Distribution")
        plt.legend(hr_employee_data[col].value_counts().index)
        plotnumber+=1
```

```python
hr_cols_2 = hr_employee_data.drop('Attrition', axis = 1).select_dtypes(['object']).columns
hr_cols_2
```

# Plotting the Bar Chart for the Attrition Distribution

```python
plt.figure(figsize=(30, 24), facecolor='white')

plotnumber = 1
for col in hr_cols_2:
    if plotnumber <= 9:
        plt.subplot(3, 3, plotnumber)
        sn.countplot(x=col, hue='Attrition', data=hr_employee_data)
        plt.title("Attrition distribution by " + col)
        ax = plt.gca()
        total_height = len(hr_employee_data['Attrition'])
        for p in ax.patches:
            percentage = '{:.1f}%'.format(100 * p.get_height() / total_height)
            x = p.get_x() + p.get_width() / 2
            y = p.get_height()
            ax.annotate(percentage, (x, y), ha='center', va='bottom')
        plotnumber += 1

plt.tight_layout()
plt.show()
```
